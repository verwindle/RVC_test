#### Что можно считать стартапом?
<img src='startup.jpg' style='width: 400px'>

Свойства “среднего” стартапа, без которых проект, идея, сайт или 
приложение не может быть причисленным к стартапу:

- стартап должен создавать новую нишу. Если говорить о проектах-двойниках, то они наполняют незаполненную нишу и удовлетворяют растущий спрос. К тому же, если рынок не растет экспоненциально, то все, что происходит в этой области должно называться бизнесом или предпринимательством;
- стартапом команда должна заниматься фуллтайм; если это не так — это хобби, бутстрепинг или еще что-то, но не стартап;
- аутсорсинг ни при каких обстоятельствах не может считаться стартапом;
- наличие бизнес-плана, сайта, приложения, уникального дизайна, инвестора, заработка, офиса, инноваций, патентов — существенное, но не обязательное условие;
- стартаперы должны быть глубо “в теме” в той области, в которой они работают, иначе — это разводняк, пиар, кумовство, откаты, случай, папа-министр, мама-красивая;
- и одно из самых важных свойств — наличие активов (это могут быть как сайты, так и приложения, патенты, знания, связи, экспертиза, каналов сбыта).


####  <font color='red'> В таком случае будем считать стартапом на рынке технологий искусственного интеллекта следующее: </font>
#### <font color='red'> Стартап  一  команда профессионалов в новой или очень молодой области (ниши рынка), основное занятие членов которой составляет самостоятельная реализация собственного оригинального проекта с целью его стремительного коммерческого распространения.  Для стартапа критичны аудитория и инвестиции. </font>

#### <font color='red'> Немаловажно - стартап должен хоть как то работать. </font>

Yva.AI это компания из Силиконовой долины, базирующаяся в Калифорнии, которая помогает организациям сохранять лучшие таланты, создавать вовлеченные команды и достигать большего. 3. #### Chatterbox
Третий стартап решил взять от себя. Несколько похоже на [Amazon echo](https://www.amazon.com/smart-home-devices/b?ie=UTF8&node=9818047011), но здесь весь фокус на детей. Клиент любопытный и очень любит игрушки. Родители любят радовать детей, а дети любят выпрашивать у родителей. Алиса от яндекса (2017), появилась значительно позже Кортаны (2009) майкрософта, но о сравнении популярности говорить не приходится.

Исследование Массачусетского технологического института, изучающее способы взаимодействия детей с разговорными помощниками, показало, что дети обычно считают их “дружелюбными” и “заслуживающими доверия"."Некоторые психологи развития, такие как Кавери Субраманьям, заведующая кафедрой детских и семейных исследований, считают, что такое восприятие “дружелюбия” может побудить детей переложить задачи на ИИ, а не выполнять их самостоятельно.

#### Почему это важно?
- Лучшая форма обучения
    - критическое мышление
    - коммуникация
    - навык решения задач
   
Компьютерный интерфейс будущего - это управляемый голосом искусственный интеллект. Миссия Chatterbox состоит в том, чтобы подготовить детей к будущему, где человекоподобный ИИ вездесущ”, - сказал Элган, основатель проекта. "Создавая аппаратное обеспечение и обучая Chatterbox, дети узнают, как и почему работает ассистентская технология, а не обманываются технологией черного ящика, чтобы думать, что голосовой ИИ является человеком или каким-то авторитетом. Я вот иногда обманываюсь технологией черного ящика, а Вы? )

- Безопасность
Использование популярных голосовых помощников растет по всему миру, по крайней мере есть Siri, Alexa, Cortana. Некоторые люди выражают опасения по поводу постоянно работающих голосовых помощников в своих домах. Отчет, подготовленный по заказу Microsoft, показал, что 41% опрошенных испытывали проблемы с доверием, в то время как треть заявили, что они предпочли бы использовать помощников, которые не используют их личную информацию или данные. 
И если во взрослом мире это известные вещи, то с детьми все серьезнее. К тому же многие родители не хотят, чтобы их  дети "учиились плохому", но дети все равно будут играться и быстро осваивать новое.
Chatterbox использует Майкрофт -  независимо разработанный и открытый помощник с открытым исходным кодом, ориентированный на конфиденциальность.

#### Технологии
В конструкторе навыков Chatterbox дети могут создавать навыки, которые усиливают обучение в классе - викторины, интерактивные исторические биографии или навыки, которые могут искать факты, такие как определения слов.

Больше можно узнать из видео.

Играя с "коробкой конструктором", дети узнают о ней - а также о ценных жизненных навыках, лежащих за пределами технологии. Chatterbox разработан, чтобы дать детям практическую среду обучения, где они могут быть творческими, развивая важные жизненные навыки, такие как эмпатия и логика.
